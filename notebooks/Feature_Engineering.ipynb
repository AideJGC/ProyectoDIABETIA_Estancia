```python
import pandas as pd

def dias_year(date):
    days = np.nan
    #print('date-',date)
    p = pd.Period(date)
    if p.is_leap_year:
        days = 366
    else:
        days = 365
        
    return days

def fecha_ini_fin(fecha_ini):
    # Crear variable de ventanas máximas o hacer referencia a la variable años de conuslta
    #fecha_ini = datetime.timestamp(fecha_ini)
    fecha_ini = pd.to_datetime(fecha_ini)
    fecha_ini = pd.Timestamp(fecha_ini)
    #print(type(fecha_ini))
    #print(fecha_ini+pd.to_timedelta(365, unit = 'D'))
    
    a_ini = fecha_ini+pd.to_timedelta(365, unit = 'D')
    a_fin = fecha_ini+pd.to_timedelta(2*365, unit = 'D')
    
    i = a_ini.strftime('%Y-%m-%d')
    f = a_fin.strftime('%Y-%m-%d')
    
    i = dias_year(i)
    #print(i)
    f = dias_year(f)
    #print(f)
    
    
    a_ini = fecha_ini+pd.to_timedelta(i, unit = 'D')
    a_fin = fecha_ini+pd.to_timedelta(i+f, unit = 'D')
    
    #print("a_ini - ",a_ini)
    #print("a_fin - ",a_fin)
    
    return a_ini, a_fin
```
